import matplotlib.pyplot as plt

explodes= [0.0, 0.1, 0.0, 0.1]
fontsize = 18
# Plot
plt.figure(figsize=(8,6))
plt.pie(sizes, labels=labels, explode=explodes, autopct='%1.1f%%', shadow=True, startangle=120, textprops={'fontsize': 10, "weight": "bold"})
plt.title(" PM characterisation for cinnamon e-liquid", fontdict={'fontsize': fontsize, "weight": "bold"},  y=1.05)

plt.show()

````python

import matplotlib.pyplot as plt

# Cinnamon flavour p<2.5

labels = ["PM < 2.21 μm", "PM < 1.72 μm", "PM < 1.47 μm","PM < 0.99 μm"]
sizes = [3572, 3200, 1313, 1468]
explodes= [0.05, 0.05, 0.05, 0.05]
fontsize = 20
# Plot
plt.figure(figsize=(8,6))
plt.pie(sizes, labels=labels, explode=explodes, autopct='%1.1f%%', shadow=True, startangle=120, textprops={'fontsize': 15, "weight": "bold"})
plt.title(" PM characterisation for cinnamon e-liquid", fontdict={'fontsize': fontsize, "weight": "bold"},  y=1.05)
plt.show()
````
