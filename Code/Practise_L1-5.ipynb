```python
sample_A = pd.DataFrame(J1)
sample_b = pd.DataFrame(J2)
```

```python
sample_A
```

```python
sample_b
```

```python
sample_inner = pd.merge(sample_A, sample_b, on = "Trips")
sample_inner
```

```python
sample_inner = pd.merge(sample_A, sample_b, on = "Trips", how= "inner")
sample_inner
```

```python
sample_outer = pd.merge(sample_A, sample_b, on = "Trips", how= "outer")
sample_outer
```
